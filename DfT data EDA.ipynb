```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import geopandas as gpd
```

# Who? When? Where? How? 

```python
data = pd.read_csv('cleaned_dft_statistics_collision_and_casualty_last_5_years.csv', low_memory = False)
data = data.copy()
```

# Who are the casualties?

```python
data.sex_of_casualty.value_counts().plot(kind='pie')
```

```python
sorter = {'0 - 5': 0, '6 - 10': 1, '11 - 15': 2, '16 - 20': 3, '21 - 25': 4, '26 - 35': 5, '36 - 45': 6, '46 - 55': 7, '56 - 65': 8, '66 - 75': 9, 'Over 75': 10}
grouped_data = data.groupby(['age_band_of_casualty', 'sex_of_casualty']).size().unstack().sort_values(by='age_band_of_casualty', key=lambda x: x.map(sorter)).plot(kind='bar')
```

```python
cas_class = data.groupby(['sex_of_casualty', 'casualty_class']).size().unstack() / data.shape[0]
ax = cas_class.plot(kind='barh')
```
